### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### Analysis
**A. Using the number of discharges as a proxy for the size of the hospital, I will explore number of discharges and excess readmission ratio **

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df['Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].apply(lambda x: int(x))
```

```python
# look at the distribution of the number of discharges
fig_d, ax_d = plt.subplots(figsize=(8,5))
x = clean_hospital_read_df['Number of Discharges']
ax_d.hist(x)
```

This histogram shows how skewed the data is, with the vast majority of hospitals have less than 1000 discharges.  This calls into question the validity of the data, since only one hospital has 7,000 discharges the sample size is much too small to make generalized statments about ALL hospitals that have 7,000 discharges.

Thus, it is important to look directly at the frequency of discharges < 2000  

```python
# determine the number of hospitals that have 2000 or more discharges
(clean_hospital_read_df['Number of Discharges'] > 2000).sum()
```

I think that given the relatively small number of hospitals with over 2000 discharges (only 32) that they should be considered separately, and not used to make generalized assumptions for the 10,000 + smaller hospitals.

```python
filtered_hospital_df = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] < 2000]
```

```python
# look at the scatterplot of the filtered data
fig_d, ax_d = plt.subplots(figsize=(8,5))
x = filtered_hospital_df['Number of Discharges']
y = filtered_hospital_df['Excess Readmission Ratio']
ax_d.scatter(x,y)
```

Based on the above scatter plot, and the previous analysis, it appears that at > 700 discharges a possible downward trend exists.  In order to test that possibility I will split the data into greater than, and less than 700 discharges, then use hypthosis testing to see if each data set has a different mean excess readmission ratio.

```python
# segment data into > 700 discharges and <= 700 discharges
large_df = filtered_hospital_df[filtered_hospital_df['Number of Discharges'] > 700]
small_df = filtered_hospital_df[filtered_hospital_df['Number of Discharges'] <= 700]
```

** Hypothesis **

H<sub>0</sub>: There is no difference between the small_df mean and large_df mean

large_df_mean - small_df_mean = 0

H<sub>1</sub>: There is a difference between the small_df mean and large_df mean

large_df_mean - small_df_mean != 0

** Decision Criteria: If the p-value result from a two-sample bootstrap hypothesis test for difference of means is < 0.05, then I will reject the null-hypothesis **


The assumption of the null hypothesis is that the large and small means are the same, so the first step is to shift both data sets so that they have the same mean as the combined dataset

```python
# determine the mean of the combined dataset
comb_mean = np.mean(filtered_hospital_df['Number of Discharges'])
# shift each dataset to the combined mean dataset
shifted_small = small_df['Number of Discharges'] - np.mean(small_df['Number of Discharges']) + comb_mean
shifted_large = large_df['Number of Discharges'] - np.mean(large_df['Number of Discharges']) + comb_mean
```

The next step is to draw 10,000 bootstrap mean replicates out of the shifted arrays and compute the difference in means.

```python
# create a function that draws bootstrap replicates
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = func(np.random.choice(data, size=len(data)))

    return bs_replicates
# draw 10,000 bootstrap samples from each shifted dateset and calculate each sample mean
large_mean_replicates = draw_bs_reps(shifted_large, np.mean, size= 10000)
small_mean_replicates = draw_bs_reps(shifted_small, np.mean, size= 10000)
# generate replicates of the difference of means
simulated_mean_diff = large_mean_replicates - small_mean_replicates
```

The last step is to compute the p-value

```python
# compute the fraction of replicates with a difference in means greater than or equal to the emperical mean
emperical_mean_diff = np.mean(large_df['Number of Discharges']) - np.mean(small_df['Number of Discharges'])
p = np.sum(simulated_mean_diff >= emperical_mean_diff)/ 10000.0
print('p-value = '+ str(p))
```

** Since p-value < 0.05, I will reject the null hypothesis and conclude that smaller hospitals have a different mean excess readmission ratio **
